```python
import json
import os
from tqdm import tqdm
```

```python
folder_path = "/data/user_data/wenkail/sotopia_diplomacy/whole_filter_games_100"
file_paths = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        file_path = os.path.join(root, file)
        file_paths.append(file_path)

games = []
num = 0
for file_path in tqdm(file_paths):
    with open(file_path, 'r') as f:
        games.append(json.load(f))
```

```python
with open("taskeval_1757_llama_with_previous.json") as f:
    task_eval = json.load(f)
```

```python
games[0]['id']
```

```python
def get_actual_orders(games, game_id, phase_name, c):
    for g in games:
        if g['id'] == game_id:
            for p in g['phases']:
                if p['name'] == phase_name:
                    orders = p['orders']
                    return orders[c]
```

```python
remains = []
for t in task_eval:
    r = {}
    r['game_id'] = t['game_id']
    r['phase_name']= t['phase_name']
    r['env_uuid'] = t['env_uuid']
    for k in t.keys():
        if "predict" in k:
            r[k] = t[k]
        if "parse" in k:
            country = k.split("_")[0]
            upper_country = country.upper()
            r[f'{country}_actual_movement'] = get_actual_orders(games, r['game_id'], r['phase_name'], upper_country)
    remains.append(r)
remains
```

```python
with open("parser_result.json", "w") as f:
    json.dump(remains, f, indent=4)
```
