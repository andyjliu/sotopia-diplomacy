```python
import json
import rich
from sotopia.database import EnvironmentProfile
```

```python
with open("taskeval_actual_demo_coop_with_actual.json", "r") as f:
    actual = json.load(f)

with open("taskeval_gpt4o_demo_coop_with_actual.json", "r") as f:
    gpt = json.load(f)
    
with open("taskeval_llama_demo_coop_with_actual.json", "r") as f:
    llama = json.load(f)
    
countries_list = ['Austria', 'England', 'France', 'Germany', 'Italy', 'Russia', 'Turkey']
```

```python
def get_score_with_env(countries, prediction):
    scores = 0
    for c in countries:
        scores += prediction[c][c.upper()][1]
    score_dict = {}
    score_dict['env'] = prediction['env_uuid']
    score_dict['scores'] = scores / len(countries)
    return score_dict

def get_score_dict_from_dialogue(dialogue):
    score_list = []
    for a in dialogue:
        keys = []
        for vkeys in a.keys():
            if vkeys in countries_list:
                keys.append(vkeys)
        score_list.append(get_score_with_env(keys, a))
    return score_list
```

```python
llama_score = get_score_dict_from_dialogue(llama)
gpt_score = get_score_dict_from_dialogue(gpt)
actual_score = get_score_dict_from_dialogue(actual)
```

```python
def caculate_score_different_via_env(model_score, actual_score):
    actual_score_dict = {item['env']: item['scores'] for item in actual_score}
    differences = []
    for litem in model_score:
        env_id = litem['env']
        if env_id in actual_score_dict:
            diff = {
                'env': env_id,
                'model_score': litem['scores'],
                'actual_score': actual_score_dict[env_id],
                'score_diff': litem['scores'] - actual_score_dict[env_id]
            }
            differences.append(diff)
    differences.sort(key=lambda x: x['score_diff'], reverse=True)
    
    return differences
```

```python
gpt_actual_diff = caculate_score_different_via_env(gpt_score, actual_score=actual_score)
llama_actual_diff = caculate_score_different_via_env(llama_score, actual_score=actual_score)
gpt_actual_diff
```
