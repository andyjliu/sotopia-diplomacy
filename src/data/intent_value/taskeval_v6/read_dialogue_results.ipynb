# Clean

```python
import json
import rich
```

```python
with open("taskeval_llama_dialogue_movement_coop_with_flausible.json", 'r') as f:
    model_with_flausible= json.load(f)    

with open("taskeval_llama_dialogue_movement_coop_without_flausible.json", 'r') as f:
    model_without_flausible = json.load(f)
```

```python
countries_list = ['Austria', 'England', 'France', 'Germany', 'Italy', 'Russia', 'Turkey']
def get_agents(phase):
    agents = []
    for key in phase.keys():
        if key in countries_list:
            agents.append(key)
    return set(agents)
```

```python
pairs = []
for wf in model_with_flausible:
    for wof in model_without_flausible:
        if wf['game_id'] == wof['game_id'] and wf['phase_name'] == wof['phase_name'] and get_agents(wf) == get_agents(wof):
            info = {
                'phase_name': wf['phase_name'],
                'game_id': wf['game_id'],
                'agents': list(get_agents(wf)),
                "dialogue_with_flausible": wf['intent_dialogue'],
                "dialogue_without_flausible": wof['intent_dialogue']
            }
            pairs.append(info)
            break
```

```python
len(model_with_flausible), len(model_without_flausible), len(pairs)
```

```python
out_file = "with_without_flausible_pairs.json"
with open(out_file, 'w') as f:
    json.dump(pairs, f, indent=4)
```
