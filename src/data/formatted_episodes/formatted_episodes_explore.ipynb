```python
import json
import rich
from tqdm import tqdm
```

```python
with open("formatted_episodes_for_specific_human_anno_gpt_4_new.json", "r") as f:
    gpt_episodes = json.load(f)

with open("formatted_episodes_for_specific_human_anno_llama3_70b.json", "r") as f:
    llama_episodes = json.load(f)
```

```python
rich.print(llama_episodes[0]['intent_dialogue'])
```

```python
# num = 0
episode_store = []
for episode_gpt in tqdm(gpt_episodes):
    for episode_llama in llama_episodes:
        if episode_gpt['env_uuid'] == episode_llama['env_uuid'] and episode_gpt['agents'] == episode_llama['agents']:
           episode_store.append([episode_gpt, episode_llama]) 
# print(num)
```

```python
# GPT Results
episode_index=1
rich.print(episode_store[episode_index][0]['dialogue'])
```

```python
# Llama Results
episode_index=1
rich.print(episode_store[episode_index][1]['dialogue'])
```
