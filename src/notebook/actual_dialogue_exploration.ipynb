```python
import random
import rich
import sys
import json
import os
sys.path.append("../../")
from tqdm import tqdm
from sotopia.database import AgentProfile, EpisodeLog, EnvironmentProfile
```

```python
folder_path = "/data/user_data/wenkail/sotopia_diplomacy/clean_global_sub_sample"
file_paths = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        file_path = os.path.join(root, file)
        file_paths.append(file_path)

games = []
num = 0
for file_path in tqdm(file_paths):
    with open(file_path, 'r') as f:
        games.append(json.load(f))
```

```python
all_character_pks = list(EnvironmentProfile.all_pks())
envs = []
for pk in all_character_pks:
    env = EnvironmentProfile.get(pk)
    if env.env_tag == "random_sample_100_games":
        envs.append(env)
```

```python
rich.print(envs[0])
```

```python
def extract_messages(messages, countries):
    countries = [country.upper() for country in countries]
    filtered_messages=[]
    for message in messages:
        if message['sender'] in countries and message['recipient'] in countries:
            filtered_messages.append(message)
    return filtered_messages
```

```python
stores = []
for env in tqdm(envs, desc = "Processing envs"):
    current_phase = env.phase_name
    current_game = env.game_id
    powers = env.agent_powers
    for game in games:
        if game['id']==current_game:
            for phase in game['phases']:
                if phase['name'] == current_phase:
                    s={}
                    s['game'] = current_game
                    s['phase'] = current_phase
                    s['messages'] = extract_messages(phase['messages'], powers)
                    s['countries'] = powers
                    stores.append(s)
    
```

```python
with open("actual_dialogue_random_sample_100_games.json", "w") as file:
    json.dump(stores, file, indent=4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

def plot_enhanced_bar_chart(list1, list2, labels):
    # 计算每个列表的平均值和标准差
    means = [np.mean(list1), np.mean(list2)]
    stds = [np.std(list1), np.std(list2)]

    # 创建自定义颜色映射
    colors = ['#1e88e5', '#FFA500']
    n_bins = 100
    cmap = LinearSegmentedColormap.from_list("custom", colors, N=n_bins)

    # 设置图表样式
    # plt.style.use('seaborn-whitegrid')
    fig, ax = plt.subplots(figsize=(12, 7))
    fig.patch.set_facecolor('#f0f0f0')  # 设置图表背景色

    # 创建条形图
    x = np.arange(len(labels))
    width = 0.4
    rects = ax.bar(x, means, width, yerr=stds, capsize=7, 
                   color=[cmap(0.2), cmap(0.8)], alpha=0.8,
                   error_kw=dict(ecolor='#333333', lw=2, capsize=7, capthick=2))

    # 设置条形图的边框和阴影效果
    for rect in rects:
        rect.set_edgecolor('#333333')
        rect.set_linewidth(1)
        plt.gca().add_patch(plt.Rectangle((rect.get_x(), 0), rect.get_width(), rect.get_height(),
                                          facecolor='none', edgecolor='none', alpha=0.3))

    # 设置图表标题和轴标签
    ax.set_title('Comparison of Two Messages', fontsize=20, fontweight='bold', pad=20)
    ax.set_xlabel('Messages Source', fontsize=14, labelpad=15)
    ax.set_ylabel('Avarage Length', fontsize=14, labelpad=15)

    # 设置x轴刻度和标签
    ax.set_xticks(x)
    ax.set_xticklabels(labels, fontsize=12)

    # 在条形上方添加数值标签
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height:.2f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=11, fontweight='bold')

    autolabel(rects)

    # 添加图例
    ax.legend(['Mean with Standard Deviation'], loc='upper right', fontsize=12)

    # 设置y轴范围，留出一些空间
    ax.set_ylim(0, max(means) * 1.2)

    # 添加网格线
    ax.grid(True, linestyle='--', alpha=0.7)

    # 移除顶部和右侧边框
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # 加粗左侧和底部边框
    ax.spines['left'].set_linewidth(2)
    ax.spines['bottom'].set_linewidth(2)

    # 调整布局并显示图表
    fig.tight_layout(pad=3.0)
    plt.show()

# 示例数据
labels = ['Actual Messages', 'Llama3 Messages']

plot_enhanced_bar_chart(len_actual_messages, len_epi_messages, labels)
```

```python
import pandas as pd
df = pd.DataFrame(stores)
df
```

```python
# Extract unique phases
unique_phases = df['phase'].nunique()

# Calculate average message length for each phase
def average_message_length(messages):
    if not messages:
        return 0
    # total_length = sum(len(message['message']) for message in messages)
    # return total_length / len(messages)
    return len(messages)

df['avg_message_length'] = df['messages'].apply(average_message_length)
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is already defined and contains the required data
grouped = df.groupby('phase')['avg_message_length'].agg(['mean', 'count'])

# Plot the statistical values
plt.figure(figsize=(10, 6))

# Plot mean with standard deviation error bars
plt.errorbar(grouped.index, grouped['mean'], fmt='o', capsize=5, capthick=2, color='red')
plt.plot(grouped.index, grouped['mean'], linestyle='-', marker='o', color='lightcoral')

plt.xlabel('Phase')
plt.ylabel('Average Message Length')
plt.title('Average Message Length per Phase')
plt.grid(True)
plt.show()
```
