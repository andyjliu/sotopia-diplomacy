```python
def get_score_with_env(countries, prediction):
    scores = 0
    for c in countries:
        scores += prediction[c][c.upper()][1]
    score_dict = {}
    score_dict['env'] = prediction['env_uuid']
    score_dict['scores'] = scores / len(countries)
    return score_dict

def get_score_dict_from_dialogue(dialogue):
    score_list = []
    for a in dialogue:
        keys = []
        for vkeys in a.keys():
            if vkeys in countries_list:
                keys.append(vkeys)
        score_list.append(get_score_with_env(keys, a))
    return score_list
```

```python
actual_dialogue_score_dict = get_score_dict_from_dialogue(human_dialogue)
actual_movement_score_dict = get_score_dict_from_dialogue(actual_movement)
llama_with_previous_score_dict = get_score_dict_from_dialogue(llama_with_previous)
llama_without_previous_score_dict = get_score_dict_from_dialogue(llama_without_previous)
# actual_dialogue_score_dict[0]
```

```python
def sub_score(score_list_1, score_list_2):
    model_score_dict = {item['env']: item['scores'] for item in score_list_2}
    results = []
    for item in score_list_1:
        env = item['env']
        score1 = item['scores']
        score2 = model_score_dict.get(env)
        if score2 == None:
            continue
        results.append({"env": item['env'], "scores": score1 - score2})
        results.append(env)
    return results
```

```python
sub_score(actual_dialogue_score_dict, llama_with_previous_score_dict)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data1 = [d['scores'] for d in llama_with_previous_score_dict]
data2 = [d['scores'] for d in llama_without_previous_score_dict]

sns.set(style="whitegrid")

# 密度图
plt.figure(figsize=(10, 5))
sns.kdeplot(data1, shade=True, color='skyblue', alpha=0.7, linewidth=2, label='Llama with Previous')
sns.kdeplot(data2, shade=True, color='green', alpha=0.7, linewidth=2, label='Llama without Previous')

# 计算并标记分位点
percentiles = [25, 50, 75, 90]
for percentile in percentiles:
    score1 = np.percentile(data1, percentile)
    plt.axvline(x=score1, color='red', linestyle='--', label=f'Dataset 1 - {percentile}th: {score1:.2f}')
    score2 = np.percentile(data2, percentile)
    plt.axvline(x=score2, color='purple', linestyle='--', label=f'Dataset 2 - {percentile}th: {score2:.2f}')

plt.title('Density: Llama With Previous vs. Llama Without Previous', fontsize=15)
plt.xlabel('Score Difference', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data1 = [d['scores'] for d in actual_dialogue_score_dict]
data2 = [d['scores'] for d in actual_movement_score_dict]

sns.set(style="whitegrid")

# 计算 90th 百分位点
p90_data1 = np.percentile(data1, 90)
p90_data2 = np.percentile(data2, 90)

# 密度图
plt.figure(figsize=(10, 5))
sns.kdeplot(data1, shade=True, color='skyblue', alpha=0.7, linewidth=2, label='Actual Dialogue')
sns.kdeplot(data2, shade=True, color='green', alpha=0.7, linewidth=2, label='Actual Movement')

# 只显示大于 90th 百分位点的部分
plt.xlim(min(p90_data1, p90_data2), max(max(data1), max(data2)))

plt.title('Density: Human Dialogue vs. Actual Movement (> 90th Percentile)', fontsize=15)
plt.xlabel('Score Difference', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
def list_choice(score_list_1, score_list_2):
    model_score_dict = {item['env']: item['scores'] for item in score_list_2}
    results = []
    for item in score_list_1:
        env = item['env']
        score1 = item['scores']
        score2 = model_score_dict.get(env)
        if score2 == None:
            continue
        if score1 >= 0.04 and score2 >= 0.06:
            results.append(env)
    return results
```
